## 1. 같은 문제, 다른 방법: 리스트에서 최댓값 찾기

같은 문제를 풀더라도 **방법(알고리즘)**에 따라 효율이 크게 달라집니다.

```python
# 방법 1: 순차 탐색 — 리스트를 한 번 훑는다 → 비교 횟수: n-1번
def find_max(numbers: list[int]) -> int:
    max_val: int = numbers[0]
    for num in numbers:
        if num > max_val:
            max_val = num
    return max_val

# 방법 2: 정렬 후 마지막 원소 가져오기 — 연산 횟수: 약 n log n번
def find_max_sort(numbers: list[int]) -> int:
    sorted_numbers: list[int] = sorted(numbers)
    return sorted_numbers[-1]

# 테스트
data = [3, 7, 2, 9, 1, 5, 8, 4, 6]
print(f"방법 1 (순차 탐색): {find_max(data)}")
print(f"방법 2 (정렬 후 선택): {find_max_sort(data)}")
print()
print("두 방법 모두 같은 결과를 내지만, 방법 1이 더 효율적입니다.")
print(f"  방법 1: 약 n번 = {len(data)}번")
print(f"  방법 2: 약 n log n번 ≈ {len(data) * 3:.0f}번")
```

## 2. 연산 횟수 세기

시간 복잡도(time complexity)란, 입력 크기 `n`이 커질 때 알고리즘이 수행하는 **기본 연산의 횟수**가 어떻게 증가하는지를 나타내는 함수입니다.

```python
# 예시 1: 합계 구하기 — T(n) = n + 2
def total(numbers: list[int]) -> int:
    result: int = 0          # 대입 1번
    for num in numbers:       # 반복 n번
        result += num         #   덧셈 1번 × n
    return result             # 반환 1번

print(total([1, 2, 3, 4, 5]))
```

```python
# 예시 2: 중복 쌍 찾기 — T(n) = n(n-1)/2 (최악의 경우)
def has_duplicate(numbers: list[int]) -> bool:
    for i, num in enumerate(numbers):        # n번
        for other in numbers[i + 1:]:        #   (n-1), (n-2), ...
            if num == other:                 #     비교 1번
                return True
    return False

print(has_duplicate([1, 2, 3, 4, 5]))   # False
print(has_duplicate([1, 2, 3, 2, 5]))   # True

# 최악의 경우 비교 횟수 확인
for n in [10, 100, 1000]:
    comparisons = n * (n - 1) // 2
    print(f"n={n:>5}: 최대 비교 횟수 = {comparisons:>10,}번")
```

## 3. 주요 복잡도 클래스별 코드 예시

### O(1): 상수 시간
입력 크기 `n`에 **관계없이** 항상 같은 시간이 걸립니다.

```python
# O(1): 리스트의 첫 번째 원소 접근
def get_first(numbers: list[int]) -> int:
    return numbers[0]

# O(1): 딕셔너리 조회
def lookup(data: dict[str, int], key: str) -> int:
    return data[key]

# 리스트가 10개이든 100만 개이든 상관없다
print(get_first([10, 20, 30]))
print(lookup({"a": 1, "b": 2}, "b"))
```

### O(n): 선형 시간
입력을 **한 번씩** 확인해야 하는 알고리즘. `n`이 2배면 시간도 2배.

```python
# O(n): 리스트에서 특정 값 찾기 (선형 탐색)
def linear_search(numbers: list[int], target: int) -> int:
    for i, num in enumerate(numbers):
        if num == target:
            return i
    return -1

# 최악의 경우: 리스트 끝까지 확인 → n번
data = [4, 7, 2, 9, 1, 5]
print(f"9의 위치: {linear_search(data, 9)}")   # 3
print(f"6의 위치: {linear_search(data, 6)}")   # -1 (없음)
```

### O(n²): 이차 시간
**이중 반복문**이 대표적. `n`이 2배면 시간은 4배.

```python
# O(n²): 모든 쌍 출력
def print_pairs(items: list[str]) -> None:
    for item1 in items:
        for item2 in items:
            print(f"({item1}, {item2})", end="  ")
        print()

# 바깥 반복문 n번 × 안쪽 반복문 n번 = 총 n²번
print_pairs(["A", "B", "C"])
```

## 4. 시간 복잡도 분석 연습

### 분석 규칙
1. **순차 실행**: 각 블록의 복잡도 중 큰 것 → `O(n) + O(n²) = O(n²)`
2. **반복문**: 반복 횟수 × 내부 복잡도 → `for문 n번 × O(1) = O(n)`
3. **중첩 반복문**: 바깥 × 안쪽 → `n번 × n번 = O(n²)`
4. **반으로 줄어드는 반복**: `O(log n)`
5. **조건문**: 더 느린 쪽의 복잡도 (최악의 경우)

```python
# 연습 1: O(n) — 두 개의 순차 for문
def example1(n: int) -> int:
    total: int = 0
    for i in range(n):
        total += i
    for j in range(n):
        total += j
    return total

# 분석: O(n) + O(n) = O(n)
print(f"example1(10) = {example1(10)}")
```

```python
# 연습 2: O(n²) — 이중 반복문 + 단일 반복문
def example2(n: int) -> int:
    total: int = 0
    for i in range(n):
        for j in range(n):
            total += i * j
    for k in range(n):
        total += k
    return total

# 분석: O(n²) + O(n) = O(n²)
print(f"example2(10) = {example2(10)}")
```

```python
# 연습 3: O(log n) — 매번 절반으로 줄어드는 반복
def example3(n: int) -> int:
    count: int = 0
    i: int = n
    while i > 0:
        count += 1
        i = i // 2
    return count

# 분석: i가 n → n/2 → n/4 → ... → 1로 줄어든다 → 반복 횟수: log₂(n)
for n in [8, 1024, 1_000_000]:
    print(f"example3({n:>10,}) = {example3(n):>3} (log₂({n}) ≈ {n.bit_length() - 1})")
```

```python
# 연습 4: O(n²) — 조건문이 있는 경우 (최악의 경우 분석)
def example4(data: list[int]) -> list[int]:
    n: int = len(data)
    result: list[int] = []
    for i in range(n):
        if data[i] > 0:
            result.append(data[i])
        else:
            for j in range(n):
                result.append(data[j])
    return result

# 분석: 최악의 경우 (모든 원소가 0 이하) → 바깥 n번 × 안쪽 n번 = O(n²)
print(f"양수 포함: {example4([1, -2, 3])}")
print(f"전부 음수: {example4([-1, -2, -3])}")
```

## 5. `timeit`으로 실행 시간 측정하기

`time.time()`은 1회 측정이라 오차가 큽니다. `timeit`은 여러 번 반복 후 합산하여 더 **정확**합니다.

```python
import timeit

# timeit 기본 사용법
elapsed: float = timeit.timeit(
    "sum(range(1000))",  # 측정할 코드 (문자열)
    number=100           # 반복 횟수
)
print(f"sum(range(1000))을 100번 실행: {elapsed:.4f}초")
```

### O(n) vs. O(n²) 비교 실험

```python
# O(n) 함수
def sum_linear(n: int) -> int:
    total: int = 0
    for i in range(n):
        total += i
    return total

# O(n²) 함수
def sum_quadratic(n: int) -> int:
    total: int = 0
    for i in range(n):
        for j in range(n):
            total += 1
    return total

# 실행 시간 비교
import timeit

for n in [1000, 5000, 10000, 20000]:
    t_lin: float = timeit.timeit(lambda: sum_linear(n), number=10)
    t_quad: float = timeit.timeit(lambda: sum_quadratic(n), number=10)
    print(f"n={n:>5}: O(n)={t_lin:.4f}초, O(n^2)={t_quad:.4f}초")
```

**관찰**
- `O(n)`: `n`이 2배 → 시간도 약 **2배**
- `O(n²)`: `n`이 2배 → 시간은 약 **4배**

## 6. 자료구조 선택의 중요성

같은 문제라도 자료구조를 잘 선택하면 알고리즘의 성능이 크게 달라집니다.

| 연산 | `list` | `dict` / `set` |
|------|--------|----------------|
| 검색 (`x in ...`) | O(n) | O(1) |

```python
# 방법 1: 리스트로 중복 검사 -- O(n^2)
def unique_list(words: list[str]) -> list[str]:
    result: list[str] = []
    for word in words:          # n번
        if word not in result:  #   최대 n번 검색
            result.append(word)
    return result

# 방법 2: 집합(set) 사용 -- O(n)
def unique_set(words: list[str]) -> list[str]:
    return list(set(words))     # set 생성 O(n)

# 테스트
sample = ["사과", "바나나", "사과", "체리", "바나나", "딸기"]
print(f"방법 1 (리스트): {unique_list(sample)}")
print(f"방법 2 (집합):   {unique_set(sample)}")
print()
print("같은 문제, 같은 결과지만 자료구조 선택으로 O(n²) → O(n)")
```

## 7. 파이썬 내장 자료형의 시간 복잡도 정리

### 리스트 (`list`)

| 연산 | 예시 | 시간 복잡도 |
|------|------|:----------:|
| 인덱스 접근 | `lst[i]` | O(1) |
| 끝에 추가 | `lst.append(x)` | O(1) |
| 끝에서 삭제 | `lst.pop()` | O(1) |
| 중간 삽입/삭제 | `lst.insert(i, x)` | O(n) |
| 검색 | `x in lst` | O(n) |
| 정렬 | `lst.sort()` | O(n log n) |

### 딕셔너리 (`dict`)

| 연산 | 예시 | 시간 복잡도 |
|------|------|:----------:|
| 키로 접근 | `d[key]` | O(1) |
| 키-값 추가 | `d[key] = val` | O(1) |
| 키 삭제 | `del d[key]` | O(1) |
| 키 검색 | `key in d` | O(1) |

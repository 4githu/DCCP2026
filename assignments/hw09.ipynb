```python
import csv
import logging
import json

logging.basicConfig(
    level=logging.INFO,
    force=True
)

def make_report(csv_path: str, json_path: str):
    try:
        dic = []
        with open(csv_path, 'r', encoding = "utf-8") as f:
            reader = csv.DictReader(f)
            for row in reader:
                #print(row)
                newdic = {}
                newdic["이름"] = row['이름']
                newdic["학번"] = row['학번']
                newdic["점수"] = {}
                isnull = 0
                for i in ['중간', '기말', '과제']:
                    if(row[i] == ''):
                        isnull = 1
                        newdic["점수"][i] = None
                    else:
                        newdic["점수"][i] = int(row[i])
                if(isnull == 1):
                    newdic["평균"] = None
                    newdic["등급"] = None
                else:
                    newdic["평균"] = (newdic["점수"]['중간']*0.3 + newdic["점수"]['기말']*0.5 + newdic["점수"]['과제']*0.2)
                    if(newdic["평균"] >= 90):
                        newdic["등급"] = 'A'
                    elif(newdic["평균"] >= 80):
                        newdic["등급"] = 'B'
                    elif(newdic["평균"] >= 70):
                        newdic["등급"] = 'C'
                    else:
                        newdic["등급"] = 'F'
                dic.append(newdic)
                logging.info(f"{newdic['이름']} : {newdic['평균']}, {newdic['등급']}")

        with open(json_path, 'w', encoding = "utf-8") as f:
            json.dump(dic, f, ensure_ascii = False, indent = 2)
    except FileNotFoundError as e:
        logging.warning(f"FileNotFoundError")
        return 0;
    except UnicodeDecodeError as e:
        logging.error(f"UnicodeDecodeError")
        return 0;



make_report('hw09files/scores.csv', 'hw09files/q1output.json')
```

기본적으로 csv 읽고, json 저장하는 코드를 짜고
json이 []으로 묶고 그 안이 {}이니 리스트 만들어서 하나씩 추가했다 (처음에 딕셔너리로 열었으니까 하고 변수명을 dic = {}으로 잡았다가 리스트로 고친..)
각 안에 넣을 딕셔너리도 추가했고 안에 있는 변수 3개 지우고 ['점수'] = {}으로 해야하는게 귀찮아서 새로 만들었다. 
비어있으면 널 넣으라고 했으니 = ''으로 판정했고 평균 구하고 등급 나눴다.
그리고 오류처리를 이어서 추가했다.

```python

class InvalidJamoError(ValueError) :
    """"""

def dtd(a):
    if(type(a) != str):
        raise TypeError()
    if(len(a) != 1):
        raise ValueError()

    if(a >= "ㄱ" and a <= "ㅎ"):
        return "자음"
    elif(a >= "ㅏ" and a <= "ㅣ"):
        return "모음"

    raise InvalidJamoError()

inputs = ["ㄱ", "ㅏ", "ㄲ", "가", "AB", 5, "ㅎ", "ㅣ", ""]
for i in inputs:
    try:
        print(dtd(i))
    except TypeError as e:
        print(f"[타입에러] 인자가 str이 아님")
    except InvalidJamoError as e:
        print(f"[자모에러] 자음이나 모음이 아님")
    except ValueError as e:
        print(f"[벨루에러] 길이가 1이 아님")
    
```

자모에러는 값이 자음이나 모음의 범위를 벗어날때를 지칭하는 에러로 결국 "자료형은 맞으나 값이 부적절한" 벨루 에러의 일종이며 자모에러를 벨루에러의 자식클래스로 잡으면 값이 범위를 벗어난 에러들을 모을때 자식클래스인 자모에러까지 빠트리지 않고 한번에 잡을 수 있기때문에 그 유용성이 더 높다.
